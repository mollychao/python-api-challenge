```python
# Dependencies
import requests
import json
import random
from citipy import citipy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Google developer API key
from config import api_key
```

```python
#create to list of randomly selected lats & lngs
lat_lngs = []
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = list(zip(lats, lngs))
```

```python
#create a list to host all the cities
cities=[]

# use citypy to find the nearest city for each latitude and longitude
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # append unique city names to city list
    if city not in cities:
        cities.append(city)
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
# set up lists to hold reponse info
lat = []
lng=[]
temp = []
humidity=[]
cloudiness=[]
wind_speed=[]
city_name=[]

#loop all the city in the list of cities

for city in cities:
    response = requests.get(query_url + city).json() 
    
    try:
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        temp.append(response['main']['temp'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        city_name.append(response['name'])
        

    except KeyError:
        print(f"Oops, {city} doesn't exist.")
```

```python
# create a data frame from cities, lat, and temp

weather_dict = {
    "City": city_name,
    "Lat": lat,
    "Lng": lng,
    "Temperature": temp,
    "Humidity": humidity,
    "Cloudiness": cloudiness,
    "Wind Speed": wind_speed
    
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```

```python
weather_data.to_csv("weather_data.csv")
```

Temperature vs. Latitude

```python
# Build a scatter plot for each data type
plt.scatter(weather_data["Lat"], weather_data["Temperature"], marker="o")

# Incorporate the other graph properties
plt.title("Temperature (C) vs. Latitude")
plt.ylabel("Temperature (C)")
plt.xlabel("Latitude")
plt.grid()

# Save the figure
plt.savefig("Temperature (C) vs. Latitude.png")

# Show plot
plt.show()
```
